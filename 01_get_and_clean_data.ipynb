```python
# Loading previously saved data

import pickle
df = pickle.load(open("kicks.pkl",'rb'))
```

```python
### Take a look at the first 5 rows of the data
df.head(5)
```

### All the data is read in into column 'data'
### It's a dictionary of dictionaries --> should be handled by JSON reader

```python
### Drop fluff
df= df['data']
```

```python
### Parse "data" into a data frame reading it as JSON
df_parsed = pd.read_json(df.to_json(orient="records"))
```

```python
df_parsed.head(3)
```

```python
### Check distribution by country (associtated with creator's location)
df_parsed.country.value_counts()
```

```python
### To keep it clean and simple for NLP analysis, stick to one language only, English
### Given that population of English-speaking countries is dominated by the U.S. and Canada, 
### the second largest host of projects, is dual-speaking, data set will be limited to US only

df_us = df_parsed.query('country == "US"').reset_index()
df_us.shape
```

```python
### Drop columns that are not going to be used in NLP/clustering analysis
df_us_ = df_us.drop(['currency_symbol',
'currency_trailing_code',
'disable_communication',
'friends',
'index',                     
'is_backing',
'is_starred',
'permissions',
'photo',
'slug',
'source_url',
'spotlight',
'static_usd_rate',
'urls'], axis=1)
```

```python
df_us_.head(5)
```

### The following columns contain dictionary of dictionaries and subject to further parsing:
-category
-creator
-location
-profile

### JSON reader still looks like a good approach

```python
parse_dict={"category":"df_category",
            "creator":"df_creator",                       
            "profile":"df_profile"
            }
```

```python
for key, value in parse_dict.items():
      parse_dict[key] = pd.read_json(df_us_[key].to_json(orient="records"))
```

### Checking the parsed data to see if anything is relevant for the purpose of the project

```python
parse_dict['category'].head(5)

### split 'slug' by "/" into 'category' and 'sub-category'
### get rid off everything else
```

```python
parse_dict['creator'].head(5)
```

```python
parse_dict['creator'].head(5)
### Keep 'name' (creator), 'is_registered'
```

```python
parse_dict['profile'].head(5)
### pretty useless (blurb is captured in 'df_parsed')
```

```python
### Location column is different -  unable to parse it like other three columns.
### Below works, however

df_location = pd.DataFrame.from_dict(df_us_.location.to_dict()).transpose()
df_location.head(5)

### grab 'country', 'name' (city),  and 'state' (state)
```

```python
### 'deadline' date is in timestamp format
### Convert to human-readable format, save as a new column, 'deadline_dt'

import datetime

def time_to_date(row):
    return datetime.datetime.fromtimestamp(row['deadline']).strftime('%Y-%m-%d %H:%M:%S')


df_us_['deadline_dt'] = df_us_.apply(time_to_date, axis =1)
```

### Now compile all the data back together
### Renaming/dropping columns as needed


```python
### Dataframe containing primary (most important) features: df_us_
### Drop columns that will be replaced with 

df_us_.drop(['category','creator','location','profile','deadline'], axis=1, inplace = True)
```

```python
### Dataframe with location of the project (doesn't always coincide with 'country' of project creator)
### Grab 'country', 'name' (city),  and 'state' (state)
df_loc = df_location[['country','name','state']]
df_loc = df_loc.rename(columns={'country':'loc_country','name':'loc_name','state':'loc_state'})
```

```python
### Dataframe with details on creator of the project 
### Grab 'name' (creator/company name(s)),  and 'is_registered'
df_crea = parse_dict['creator'].drop(['avatar','id','slug','urls'],axis=1).rename(columns={'name':'creator_name',
                                                                                 'is_registered':'creator_registered'})
```

```python
### Dataframe with details on category of the project 
### Split 'slug' into 'category' and 'subcategory'
parse_dict['category']['category_'] = parse_dict['category']['slug'].str.split('/')
parse_dict['category']['category'] =parse_dict['category']['category_'].str[0]
parse_dict['category']['subcategory'] = parse_dict['category']['category_'].str[1]

### Get rid off other columns
df_cat = parse_dict['category'][['category','subcategory']]
```

```python
### Check that the shapes of all dataframes are as expected
print(df_cat.shape)
print(df_crea.shape)
print(df_loc.shape)
print(df_us_.shape)
```

```python
### Since order of rows has been intact on all 4 dataframes,
### joining them row by row into one final 

df_final = pd.concat([df_us_, df_cat, df_crea, df_loc], axis=1, join='inner')
```

```python
df_final.head(3)
```

```python
### Check country of project location (as opposed to creator location) distribution
df_final.loc_country.value_counts()
```

```python
### For cleaner results, drop non-US-located projects

df_final_ = df_final.query('loc_country == country').reset_index().drop('index',axis=1)
print(df_final_.head(5))
print(df_final_.shape)
```

```python
### Checking distribution of project outcomes ('state')
df_final_.state.value_counts()
```
